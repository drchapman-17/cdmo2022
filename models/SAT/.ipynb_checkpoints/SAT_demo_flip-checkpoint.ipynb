```python
def write_file(W, H, n, w, h, x, y, out_file):
    with open(out_file, 'w+') as f_out:
        f_out.write('{} {}\n'.format(W, H))
        f_out.write('{}\n'.format(n))

        for i in range(n):
            f_out.write('{} {} {} {}\n'.format(w[i], h[i], x[i], y[i]))


def read(filename):
    with open(filename) as file:  # Use file to refer to the file object
        first_line = file.readline().strip().split(" ")

        width = int(first_line[0])
        height = int(first_line[1])

        n = int(file.readline().strip())

        # Read all the remaining lines which contains the horizontal and vertical dimension of the i-th circuit
        # and its bottom left corner coordinate
        remaining_lines = file.readlines()

        # To remove \n
        remaining_lines = [line.strip() for line in remaining_lines if line.strip()]

        circuits = []
        solution = {'corners': []}

        for i in range(n):
            line = remaining_lines[i]
            line = line.split()
            circuits.append((int(line[0]), int(line[1])))
            solution['corners'].append((int(line[2]), int(line[3])))
        # Solution
    sol = {
        "w": width,
        "h": height,
        "n": n,
        "points" : np.array(circuits,\
                            dtype=[('w','i4'),('h','i4')])
    }


    print(sol['points'])
    print([(x+1,y+1) for (x,y) in solution['corners']])

    a = sol['points']
    b = solution['corners']

    l = [[sol['w'], sol['h']]]
    for i in range(len(a)):
        l.append([a[i][0],a[i][1],b[i][0]+1,b[i][1]+1])
    return l 
```

```python
import numpy as np
instn=1 #SET AS YOU PREFER
instance=utils.loadInstance(f"../instances/ins-{instn}.txt")
print("Instance:",*["{}={}".format(key,val) for key,val in instance.items()])

n=instance['n']
W=instance['w']
widths=[i[0] for i in instance['dim']]
heights=[i[1] for i in instance['dim']]

Hmax=sum(heights)
tot_area=sum([w*h for w,h in zip(widths,heights)])

lb = int(tot_area/W)
ub = int(2*max( 
            max([ [p[0],p[1]] for p in instance['dim']],key=lambda p:p[0]*p[1]) [1] #Height of the block with the largest ares
            ,lb))+1 
print('lb', lb)
print('ub', ub)

H = lb

while (lb < ub):
    s = Solver()
    px = [[Bool(f"px{i+1}_{x}") for x in range(W)] for i in range(n)] # x[i][i]: block i has y coordinate <=j
    py = [[Bool(f"py{i+1}_{y}") for y in range(H)] for i in range(n)] # x[i][i]: block i has y coordinate <=j
    

    flip= [Bool(f"f{i+1}_{y}") for i in range(n)] Presto per usarla
    px_f = [[Bool(f"px_f{i+1}_{x}") for y in range(H)] for i in range(n)] # x[i][i]: block i has y coordinate <=j
    py_f = [[Bool(f"px_f{i+1}_{y}") for x in range(W)] for i in range(n)] # x[i][i]: block i has y coordinate <=j
    
    
    lr = [[Bool(f"lr_{i+1}_{j+1}") if j!=i else 0 for j in range(n)] for i in range(n)]
    ud = [[Bool(f"ud_{i+1}_{j+1}") if i!=j else 0 for j in range(n)] for i in range(n)]

    
    # Exactly One constraint
    for i in range(n): 
        s.add(And([px[i][j] for j in range(W-widths[i],W)])) # block i starts for sure before position W-wi-> x[i][j]=True for all j > W-wi 
        s.add(And([py[i][j] for j in range(H-heights[i],H)])) # same for the y coordinate 
        s.add(And([Or(Not(px[i][j]),px[i][j+1]) for j in range(W-1)])) #x[i][j] implies x[i][j+1]
        s.add(And([Or(Not(py[i][j]),py[i][j+1]) for j in range(H-1)])) #y[i][j] implies y[i][j+1]

        for j in range(n):
            if i<j:
                s.add(Or(lr[i][j],lr[j][i],ud[i][j],ud[j][i]))
            if i!=j:    
                s.add(Or(
                        Not(lr[i][j]),
                        Not(px[j][widths[i]-1])
                    )) #under(ri,rj)-> yj > hi, lower bound for yj
                
                for e in range(0,W-widths[i]):
                    s.add(Or(
                        Not(lr[i][j]),
                        px[i][e],
                        Not(px[j][e+widths[i]])
                    ))
                
                s.add(Or(
                        Not(ud[i][j]),
                        Not(py[j][heights[i]-1])
                    )) #under(ri,rj)-> yj > hi, lower bound for yj
                
                for f in range(0,H-heights[i]):
                    s.add(Or(
                        Not(ud[i][j]),
                        py[i][f],
                        Not(py[j][f+heights[i]])
                    ))
                
    if s.check() == sat:
        ub = H
        print('SAT:ub->',ub)
    else:
        lb = H + 1
        print('UNSAT:lb->',lb)
        
    H =  int((lb+ub)/2)
# print (s.model(), px, py, H)

def converter_sat_coord(m, x, y, W, H, n):
    x_sol = []
    y_sol = [] 
    
    for i in range(n):
        j = 0
        while j < W:
            if m.evaluate(x[i][j]):
                x_sol.append(j)
                break
            j += 1

        j = 0
        while j < H:
            if m.evaluate(y[i][j]):
                y_sol.append(j)
                break
            j += 1

    return [x_sol, y_sol]


m=s.model()
out_file="DIOC"
cx, cy = converter_sat_coord(m,px,py,W,H,n)
write_file(W, H, n, widths,
               heights, cx, cy, out_file)
l=read(out_file)
utils.show(l,figsize=(15,15))

```
